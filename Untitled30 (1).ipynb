```python
print ("División en conjunto de entrenamiento y prueba")
df = pd.read_csv('heart.csv')

df = df.drop('target', axis=1)
df.head()
print ("primera parte de este codigo")

print("dividir")
from sklearn.model_selection import train_test_split
df = pd.read_csv('heart.csv')
y= df["target"]
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=42)
print(f"tamaño de entrenamiento: {len(X_train)}")
print(f"tamaño de prueba: {len(X_test)}")
```

```python
print ("Entrenamiento del modelo")

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

modelo = RandomForestClassifier(n_estimators=100, random_state=42)
modelo.fit(X_train, y_train)

y_pred = modelo.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accurary = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accurary}")
accuracy_score(y_test, y_pred)
print(f"Accuracy: {accurary}")
```
